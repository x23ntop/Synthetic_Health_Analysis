```python
from datetime import date

# Funci√≥n para calcular la edad exacta
def calcular_edad(fecha_nac):
    hoy = date.today()
    return hoy.year - fecha_nac.year - ((hoy.month, hoy.day) < (fecha_nac.month, fecha_nac.day))

# Aplicar la funci√≥n
df['Edad_calculada'] = df['Fecha Nacimiento'].apply(lambda x: calcular_edad(x))

# Comparar la edad declarada con la calculada
df['Diferencia_edad'] = df['EDAD'] - df['Edad_calculada']

# Mostrar casos con diferencias importantes
diferencias_edad = df[abs(df['Diferencia_edad']) > 1]
print(diferencias_edad[['Fecha Nacimiento', 'EDAD', 'Edad_calculada', 'Diferencia_edad']])

# Verificaci√≥n de grupo etario
def obtener_grupo_etario(edad):
    if edad < 20:
        return '<20'
    elif edad < 30:
        return '20-29'
    elif edad < 40:
        return '30-39'
    elif edad < 50:
        return '40-49'
    elif edad < 60:
        return '50-59'
    else:
        return '60+'

df['Grupo_calculado'] = df['Edad_calculada'].apply(obtener_grupo_etario)

# Mostrar inconsistencias en grupo etario
grupo_etario_inconsistente = df[df['Grupo Etareo'] != df['Grupo_calculado']]
print(grupo_etario_inconsistente[['EDAD', 'Edad_calculada', 'Grupo Etareo', 'Grupo_calculado']])
```

```python
# Agregar columnas de edad calculada y grupo etario corregido

# Calcular la edad de forma correcta (Ajusta a√±o bisiesto)
df['Edad_calculada'] = ((pd.Timestamp.today() - df['Fecha Nacimiento']).dt.days / 365.25).astype(int)

# Funci√≥n para asignar grupo etario calculado
def calcular_grupo_etario(edad):
    if edad >= 20 and edad < 30:
        return '20-29'
    elif edad >= 30 and edad < 40:
        return '30-39'
    elif edad >= 40 and edad < 50:
        return '40-49'
    elif edad >= 50 and edad < 60:
        return '50-59'
    else:
        return '60+'

# Crear la columna con grupo etario calculado
df['Grupo_calculado'] = df['Edad_calculada'].apply(calcular_grupo_etario)

```

```python
# Manejo valores Unknown

# Convertir los valores 'Unknown' a NaN en las columnas relevantes
df['hemo'] = df['hemo'].replace('unknown', pd.NA)
df['Profesion'] = df['Profesion'].replace('unknown', pd.NA)

# Imputar 'Area' con la moda
df['Area'] = df['Area'].replace('unknown', df['Area'].mode()[0])

# Imputar 'Frec_alcohol' con la moda
df['Frec_alcohol'] = df['Frec_alcohol'].replace('unknown', df['Frec_alcohol'].mode()[0])

# Imputar 'Tipo_actifisica' basado en la columna 'Actifisica'
df['Tipo_actifisica'] = df.apply(lambda row: 'regular' if row['Actifisica'] == 1.0 else row['Tipo_actifisica'], axis=1)
df['Tipo_actifisica'] = df.apply(lambda row: 'no realiza' if row['Actifisica'] == 0 else row['Tipo_actifisica'], axis=1)

# Imputar 'Frec_actifisica' basado en 'Tipo_actifisica'
df['Frec_actifisica'] = df.apply(lambda row: 'no realiza' if row['Tipo_actifisica'] == 'no realiza' else row['Frec_actifisica'], axis=1)

# Imputar 'SignosVitales_Dominancia' con la moda
df['SignosVitales_Dominancia'] = df['SignosVitales_Dominancia'].replace('unknown', df['SignosVitales_Dominancia'].mode()[0])

# Imputar 'SignosVitales_Contextura' basado en 'IMC'
df['SignosVitales_Contextura'] = df.apply(lambda row: 'delgado(a)' if row['SignosVitales_IMC'] < 18.5 else 'robusto(a)' if row['SignosVitales_IMC'] >= 30 else 'medio(a)', axis=1)


# Imputar 'SignosVitales_Tensionarterialsistolica_interpretacion' en funci√≥n de los valores de 'SignosVitales_Tensionarterialsistolica'
df['SignosVitales_Tensionarterialsistolica_interpretacion'] = df['SignosVitales_Tensionarterialsistolica'].apply(
    lambda x: 'normal' if x < 120 else 'prehipertension' if 120 <= x < 140 else 'hipertension'
)

# Imputar 'SignosVitales_Tensionarterialdiastolica_interpretacion' en funci√≥n de los valores de 'SignosVitales_Tensionarterialdiastolica'
df['SignosVitales_Tensionarterialdiastolica_interpretacion'] = df['SignosVitales_Tensionarterialdiastolica'].apply(
    lambda x: 'Normal' if x < 80 else 'Prehipertension' if 80 <= x < 90 else 'Hipertension'
)

# Dejar 'DX1_examen', 'DX2_examen', 'DX3_examen' como NaN si son 'Unknown'
df['DX1_examen'] = df['DX1_examen'].replace('unknown', pd.NA)
df['DX2_examen'] = df['DX2_examen'].replace('unknown', pd.NA)
df['DX3_examen'] = df['DX3_examen'].replace('unknown', pd.NA)

# Verificar si quedan valores nulos despu√©s de la imputaci√≥n
print(df.isnull().sum())
```

```python
# Inconsistencias en Talla y Peso

# Filtrar valores inconsistentes en 'Talla' (menos de 140 cm o m√°s de 220 cm)
df['SignosVitales_Talla'] = df['SignosVitales_Talla'].apply(lambda x: pd.NA if x < 140 or x > 220 else x)

# Filtrar valores inconsistentes en 'Peso' (menos de 30 kg o m√°s de 150 kg)
df['SignosVitales_Peso'] = df['SignosVitales_Peso'].apply(lambda x: pd.NA if x < 30 or x > 150 else x)

print(df[['SignosVitales_Talla', 'SignosVitales_Peso']].head(10))

print(df[['SignosVitales_Talla', 'SignosVitales_Peso']].isnull().sum())
```

```python
# Imputar la media de Talla y Peso segun Genero (F y M) y Sede (Baq, Cal y Bog)

# Imputar Talla basada en la media por g√©nero y sede
df['SignosVitales_Talla'] = df.groupby(['Genero', 'Sede'])['SignosVitales_Talla'].transform(lambda x: x.fillna(x.mean()))

# Imputar Peso basada en la media por g√©nero y sede
df['SignosVitales_Peso'] = df.groupby(['Genero', 'Sede'])['SignosVitales_Peso'].transform(lambda x: x.fillna(x.mean()))

# Verificar los primeros valores despu√©s de la imputaci√≥n
print(df[['SignosVitales_Talla', 'SignosVitales_Peso']].head(10))

# Verificar cu√°ntos valores NaN hay despu√©s de la imputaci√≥n
print(df[['SignosVitales_Talla', 'SignosVitales_Peso']].isnull().sum())
```

```python
# Nuevo C√°lculo de IMC basado en Talla (cm) y Peso (kg) 

df['IMC_calculado'] = df['SignosVitales_Peso'] / (df['SignosVitales_Talla'] / 100) ** 2

print(df[['SignosVitales_Talla', 'SignosVitales_Peso', 'IMC_calculado']].head(10))
```

```python
# Clasificaci√≥n IMC

def clasificar_imc(imc):
    if imc < 18.5:
        return 'Delgado aceptable'
    elif 18.5 <= imc < 25:
        return 'Normal'
    elif 25 <= imc < 30:
        return 'Sobrepeso'
    elif imc >= 30:
        return 'Obeso Tipo 1'
    else:
        return 'Sin clasificaci√≥n'

df['IMC_clasificacion'] = df['IMC_calculado'].apply(clasificar_imc)
```

```python
# Riesgo de Salud Global basado en (Genero, Edad, IMC, Actividad Fisica, D√≠as Perdidos y D√≠as de Atenci√≥n m√©dica)

def calcular_riesgo(row):
    riesgo = 0

    # Edad y g√©nero
    if (row['Genero'] == 'F' and row['Edad_calculada'] > 60) or (row['Genero'] == 'M' and row['Edad_calculada'] > 55):
        riesgo += 1
    elif (row['Genero'] == 'F' and 45 <= row['Edad_calculada'] <= 60) or (row['Genero'] == 'M' and 40 <= row['Edad_calculada'] <= 55):
        riesgo += 0.5

    # IMC
    if row['IMC_calculado'] > 30:
        riesgo += 1
    elif 25 <= row['IMC_calculado'] <= 30:
        riesgo += 0.5

    # D√≠as perdidos
    if row['Dias Perdidos'] > 5:
        riesgo += 1
    elif 2 <= row['Dias Perdidos'] <= 5:
        riesgo += 0.5

    # D√≠as de atenci√≥n m√©dica
    if row['Atenciones'] > 4:
        riesgo += 1

    # Actividad f√≠sica
    if row['Actifisica'] == 0:
        riesgo += 1
    elif row['Frec_actifisica'] in ['1 vez a la semana', 'Ocasional']:
        riesgo += 0.5

    return riesgo

# Aplicar funci√≥n
df['Riesgo_Salud_Global'] = df.apply(calcular_riesgo, axis=1)

# Clasificar riesgo
def clasificar_riesgo(valor):
    if valor <= 1:
        return 'Bajo'
    elif 1.5 <= valor <= 2.5:
        return 'Moderado'
    else:
        return 'Alto'

df['Clasificacion_Riesgo'] = df['Riesgo_Salud_Global'].apply(clasificar_riesgo)
```

```python
# Parte 2: Visualizaciones

import matplotlib.pyplot as plt
import seaborn as sns

# Relaci√≥n entre IMC y Actividad F√≠sica
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='IMC_clasificacion', y='Frec_actifisica', palette='coolwarm')
plt.title('Relaci√≥n entre IMC y Frecuencia de Actividad F√≠sica')
plt.xlabel('Categor√≠a IMC')
plt.ylabel('Frecuencia de Actividad F√≠sica')
plt.show()

# Relaci√≥n entre IMC, Edad y Riesgo
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='Edad_calculada', y='IMC_calculado', hue='Riesgo_Salud_Global', palette='coolwarm')
plt.title('Relaci√≥n entre Edad, IMC y Riesgo de Salud')
plt.xlabel('Edad')
plt.ylabel('IMC')
plt.legend(title='Riesgo de Salud')
plt.show()

# Distribuci√≥n de Riesgo por Sede
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Riesgo_Salud_Global', hue='Sede', palette='Blues')
plt.title('Distribuci√≥n de Riesgo de Salud Global por Sede')
plt.xlabel('Nivel de Riesgo')
plt.ylabel('Cantidad de Personas')
plt.legend(title='Sede')
plt.show()

# Gr√°fico de distribuci√≥n de riesgo por G√©nero
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Riesgo_Salud_Global', hue='Genero')
plt.title('Distribuci√≥n de Riesgo de Salud Global por G√©nero')
plt.xlabel('Nivel de riesgo')
plt.ylabel('Cantidad de personas')
plt.show()

```
