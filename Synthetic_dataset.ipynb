```python
# Parte 1: An√°lisis Exploratorio

import pandas as pd

df = pd.read_csv('clean_synthetic_health_data.csv', encoding='utf-8-sig',)

# ‚úÖ 1.1 Ver tama√±o y primeras filas del dataset
print("Forma del DataFrame (filas, columnas):", df.shape)
display(df.head())

# ‚úÖ 1.2 Tipos de datos y estructura general
print("\nüìä Informaci√≥n general del DataFrame:")
df.info()

# ‚úÖ 1.3 Revisi√≥n de valores nulos
print("\nüîé Cantidad de valores nulos por columna:")
display(df.isnull().sum().sort_values(ascending=False))

# ‚úÖ 1.4 Revisi√≥n de valores 'Unknown'
print("\nüîé Cantidad de valores 'Unknown' por columna:")
unknown_counts = (df == 'Unknown').sum().sort_values(ascending=False)
display(unknown_counts[unknown_counts > 0])  # Solo muestra columnas que tengan 'Unknown'

# ‚úÖ 1.5 Resumen estad√≠stico de columnas num√©ricas
print("\nüìà Resumen estad√≠stico de variables num√©ricas:")
display(df.describe().transpose())

# ‚úÖ 1.6 Valores √∫nicos en columnas categ√≥ricas clave
columnas_categoricas = [
    'Genero', 'Estado civil', 'Escolaridad', 
    'Fuma', 'bebealcohol', 'Actifisica', 'Sede'
]

for col in columnas_categoricas:
    print(f"\n‚û°Ô∏è Valores √∫nicos en la columna '{col}':")
    print(df[col].unique())

```

```python
import pandas as pd

# Carga el DataFrame normalmente
df = pd.read_csv('clean_synthetic_health_data.csv', encoding='utf-8')

# Ahora revisa el archivo l√≠nea por l√≠nea:
with open('clean_synthetic_health_data.csv', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# Cantidad de columnas que deber√≠a tener cada fila (las del DataFrame)
expected_columns = len(df.columns)
print(f"El n√∫mero esperado de columnas es: {expected_columns}")

# Revisar filas que no coincidan
for i, line in enumerate(lines):
    columns_in_line = len(line.strip().split(','))
    if columns_in_line != expected_columns:
        print(f"‚ö†Ô∏è L√≠nea {i+1} tiene {columns_in_line} columnas (deber√≠a tener {expected_columns})")

print("‚úÖ Revisi√≥n completada.")
```

```python
# Revisar y mostrar las l√≠neas problem√°ticas
for i, line in enumerate(lines):
    columns_in_line = len(line.strip().split(','))
    if columns_in_line != expected_columns:
        print(f"\nL√≠nea {i+1} con {columns_in_line} columnas:")
        print(line)
```
